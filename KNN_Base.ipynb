```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
data=pd.read_csv("diabetes.csv")
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.describe()
```

```python
X=data.iloc[:, 0:8]
X.shape
```

```python
Y=data['Diabetes']
Y.shape
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
knn=KNeighborsClassifier(n_neighbors=5, metric="euclidean")
cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
accuracy_scores_knn=cross_val_score(knn, X_scaled, Y, cv=cv, scoring="accuracy")
```

```python
print("Accuract de cada bloque", accuracy_scores_knn)
print("Accuract promedio", np.mean(accuracy_scores_knn))
print("Accuracy desviacion", accuracy_scores_knn.std())
```

```python
from sklearn.metrics import f1_score
f1_scores_knn=cross_val_score(knn, X_scaled, Y, cv=cv, scoring="f1")
```

```python
print("F1 de cada bloque", f1_scores_knn)
print("F1 promedio", f1_scores_knn.mean())
print("F1 desviacion", f1_scores_knn.std())
```
